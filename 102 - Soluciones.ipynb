### virtualenvs

1. Crear un entorno virtual usando el módulo 'virtualenvs' para python 3.8. Usa pyenv
$ pyenv install 3.8.0
$ virtualenvs myenv38 --python=python3.8

2. Crear un entorno virtual usando el módulo 'virtualenvs' para python 3.10. Usa pyenv
$ pyenv install 3.10.0
$ virtualenvs myenv310 --python=python3.10


3. Instalar un paquete en el entorno virtual python 3.8 usando 'pip'. Recuerda que el entorno este activo. Puedes tomar el nombre de algún paquete que nombramos en esta guía.
$ source myenv38/bin/activate
$ pip install numpy



4. Instalar un paquete con una versión específica en el entorno virtual python 3.10 usando 'pip'. Recuerda que el entorno este activo. Puedes tomar el nombre de algún paquete que nombramos en esta guía.
$ source myenv310/bin/activate
$ pip install numpy==1.18.5

5. Crear un archivo requirements.txt con los paquetes instalados en el entorno virtual.
pip freeze > requirements.txt

### Docker
1. Crear un contenedor de Docker con una imagen de Python 3.1 y ejecutar un script de Python que imprima "hello world" en la terminal.

Para crear el contenedor, primero debemos crear un archivo llamado Dockerfile con el siguiente contenido:

``` Dockerfile
FROM python:3.1
COPY script.py /script.py
CMD python /script.py
```

Luego, creamos un archivo script.py con el siguiente contenido:

``` script.py
print("hello world")
```

Para construir la imagen, nos ubicamos en el mismo directorio donde se encuentran los archivos Dockerfile y script.py y ejecutamos el siguiente comando:

``` bash
docker build -t my-python-app .
```

Finalmente, podemos ejecutar el contenedor con el siguiente comando:

``` bash
docker run my-python-app
```

2. Usar un archivo Dockerfile para construir una imagen personalizada de Python que incluya una librería adicional y ejecutar un script de Python que utilice dicha librería.

``` Dockerfile
FROM python:3.10
RUN pip install numpy
COPY script.py /script.py
CMD python /script.py
```

``` script.py
import numpy as np
print(np.random.rand(5))
```

``` bash
docker build -t my-python-app .
docker run my-python-app
```

3. Crear un contenedor de Docker con una imagen de Python y ejecutar un script de Python que interactúe con una API externa.

Para este ejercicio, supongamos que queremos consumir una API de prueba que devuelve un objeto JSON con información sobre un libro. Primero creamos un archivo Dockerfile con el siguiente contenido:

``` Dockerfile
FROM python:3.10
RUN pip install requests
COPY script.py /script.py
CMD python /script.py
```

``` script.py
import requests
response = requests.get('https://jsonplaceholder.typicode.com/todos/1')
print(response.json())
```

``` bash
docker build -t my-python-app .
docker run my-python-app
```
